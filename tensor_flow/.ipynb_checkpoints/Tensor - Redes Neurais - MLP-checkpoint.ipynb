```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
# Definição dos inputs e outputs

X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
], dtype=np.float64)

y = np.array([
    [0],
    [1],
    [1],
    [0]
], dtype=np.float64)
```

```python
neuronios_entrada = 2
neuronios_camada_oculta = 3
neuronios_saida = 1
```

```python
pesos = {
    'oculta': tf.Variable(
        tf.random_normal([neuronios_entrada,
                          neuronios_camada_oculta], dtype=tf.float64)),
    'saida': tf.Variable(
        tf.random_normal([neuronios_camada_oculta,
                         neuronios_saida], dtype=tf.float64))
}

bias = {
    'oculta' : tf.Variable(tf.random_normal([neuronios_camada_oculta], dtype=tf.float64)),
    'saida' : tf.Variable(tf.random_normal([neuronios_saida], dtype=tf.float64))
}
```

```python
# recebendo os dados atraves do placeholder

xph = tf.placeholder(tf.float64, [None, neuronios_entrada])
yph = tf.placeholder(tf.float64, [None, neuronios_saida])
```

```python
# Calculo para os neuronios da camada oculta

#Feed-foward

oculta = tf.add(bias['oculta'], tf.matmul(xph, pesos['oculta']))
oculta = tf.sigmoid(oculta)
saida = tf.add(bias['saida'], tf.matmul(oculta, pesos['saida']))
saida = tf.sigmoid(saida)

# Calculando o Erro
# yph são as variaveis 'target' de treinamento, portanto, são o 'ground-true'
# saida são as variaveis 'target' que foram calculadas pela rede neural.
erro =  tf.losses.mean_squared_error(yph, saida)

#Otimizador
otimizador = tf.train.AdamOptimizer(learning_rate=0.01).minimize(erro)
```

```python
class MLP_One_Layer():
    """
    Implementação de uma Rede Neural Densa com um Hidden Layer.
    """
    def __init__(self, entrada, hidden, saida):
        self.entrada = entrada
        self.hidden = hidden
        self.saida = saida
        # inicializa os pesos e bias
        self.weights = {
            'input-hidden': tf.Variable(tf.random_normal([self.entrada, self.hidden], dtype=tf.float64)),
            'hidden-output': tf.Variable(tf.random_normal([self.hidden, self.saida], dtype=tf.float64))            
        }
        self.bias = {
            'input-hidden': tf.Variable(tf.random_normal([self.hidden], dtype=tf.float64)),
            'hidden-output': tf.Variable(tf.random_normal([self.saida], dtype=tf.float64))  
        }
        self.otimizador = tf.train.AdamOptimizer(learning_rate=0.001).minimize(self.erro)
        
    @property
    def erro(self, x, y):
        """
        Passos para implementação da Rede neural Densa.
        """
        
        oculta = tf.add(bias['input-hidden'], tf.matmul(x, self.weights['input-hidden'])) #multiplicação e adição do bias
        oculta = tf.sigmoid(oculta) # valores dos neuronios. Esse vetor tem o tamanho [self.hidden]
        saida = tf.add(bias['hidden-output'], tf.matmul(oculta, self.weights['hidden-output'])) # Layer Oculto multiplicado pelos pesos para gerar a saída
        saida = tf.sigmoid(saida) # valor dos neuronios do output. 
        
        self.erro = tf.losses.mean_squared_error(y, saida)
        
        
    def train(self, inputs, target, epochs=10000):
        init = tf.global_variables_initializer()
        with tf.Session() as sess:
            sess.run(init)
            
            for epoca in range(epochs):
                erro_medio = 0
                _, loss = sess.run([self.otimizador, self.erro])
```

```python
init = tf.global_variables_initializer()
epocas = 10000
with tf.Session() as sess:
    sess.run(init)

    # definição do treinamento
    
    for epoca in range(epocas):
        erro_medio = 0
        _, custo = sess.run([otimizador, erro], feed_dict={xph: X, yph:y})
        if epocas % 200 == 0:
            erro_medio = custo/X.shape[0]
            print("MSE: ", erro_medio)
        if custo == 0:
            break
    w_final, b_final = sess.run([pesos, bias])
```
