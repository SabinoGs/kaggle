# LSTM

```python
# definição das entradas e label

periodos = 30 #intevalo de tempo. (Nos ultimos 30, preveja os proximos X*previsao_futura)
previsao_futura = 1 # horizonte

features = data[['Visibility', 'RelativeHumidity']].values
labels = data['Altimeter'].values

X_train,X_test,y_train,y_test = train_test_split(features, labels, test_size=0.2, shuffle=False, random_state=42)
```

```python
tf.reset_default_graph()

n_features = 2
neuronios_camada_oculta = 100
neuronios_saida = 1
batch_size = 30
epochs = 200


#Placeholders
xph = tf.placeholder(tf.float32, [None, n_features])
yph = tf.placeholder(tf.float32, None)
```

```python
def rnn_model():
    layer = {
        'weights': tf.Variable(tf.random_normal([neuronios_camada_oculta, neuronios_saida])),
        'bias': tf.Variable(tf.random_normal([neuronios_saida]))
    }
    
    x = tf.split(xph, n_features, 1)
    
    lstm_cell = tf.nn.rnn_cell.LSTMCell(neuronios_camada_oculta, activation=tf.nn.relu)
    out, states = tf.nn.static_rnn(lstm_cell, x, dtype=tf.float32)
    output = tf.matmul(out[-1], layer['weights']) + layer['bias']
    
    return output
```

```python
def train_nn():
    
    _batch = int(len(X_train)/batch_size)
    
    logit = rnn_model()
    logit = tf.reshape(logit, [-1])
    
    cost = tf.losses.mean_squared_error(labels=logit, predictions=yph)
    optimizer = tf.train.AdamOptimizer(learning_rate=0.001).minimize(cost)
    
    with tf.Session() as sess:
        tf.global_variables_initializer().run()
        tf.local_variables_initializer().run()
        
        for epoch in range(epochs):
            epoch_loss = 0
            i = 0
            
            for i in range(_batch):
                start = i
                end = i + batch_size
                
                batch_x = np.array(X_train[start:end])
                batch_y = np.array(y_train[start:end])
                
                _, c = sess.run([optimizer, cost], feed_dict={
                    xph: batch_x,
                    yph: batch_y
                })
                
                epoch_loss += c
                i += batch_size
            
            print('Epoch ', epoch, '/',epochs, ' Loss: ', epoch_loss)
        pred = tf.round(tf.nn.sigmoid(logit)).eval({xph: X_test, yph: y_test})
        
train_nn()
```
